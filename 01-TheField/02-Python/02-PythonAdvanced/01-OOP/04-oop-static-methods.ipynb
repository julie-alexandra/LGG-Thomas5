# What is a static method in python?

Static methods are methods within a class that have no access to anything else in the class (no `self` keyword or `cls` keyword). 

- They cannot change or look at any object attributes or call other methods within the class. 
- They can be thought of as a special kind of function that sits inside of the class. 
- When we create a static method we must use something called a [decorator](https://realpython.com/primer-on-python-decorators/#syntactic-sugar). The decorator for a static method is `@staticmethod`. No worries, we'll go deeper into the decorators later.

In other words, you can create a callable class using the static method and use it with some restrictions. It helps developers write code in a safe architectural way to prevent conflicts in the code.

Static methods in Python are used for several reasons and in specific scenarios:

- Utility functions: Static methods are ideal for creating utility functions that don't require access to instance or class state. They can perform operations that are related to the class but don't need to modify or access class-specific data.
- Code organization: Static methods allow you to group related functionality together within a class, improving code organization and readability.
- Namespace management: They provide a way to keep functions that are related to a class within its namespace without requiring an instance of the class
- Performance: Static methods can be slightly faster than instance methods because they don't need to handle the implicit self argument
- Independence from state: When a method's behavior should not depend on object or class state, making it static clearly communicates this independence
- Inheritance considerations: Static methods can be inherited and used in subclasses without modification, which can be useful in creating extensible class hierarchies

Use static methods when:

- The functionality is related to the class but doesn't need access to class-specific data.
- You want to perform operations that are independent of class and instance variables.
- You need utility functions that make sense to be part of the class but don't require an object's state.
- You want to clearly signal that a method won't modify class or instance state

Remember, static methods cannot access or modify class state or instance state directly. If you need to work with class-specific data, consider using class methods instead5

### Example: Calculator Class

Below we definte the `Calculator` class with two methods `multiply()` and `add()` which is static.

```python
class Calculator:
    """
    Class that contains methods to perform basic operations.
    """
    def __init__(self, number_one, number_two):
        self.number_one = number_one
        self.number_two = number_two

    def multiply(self):
        result = self.number_one * self.number_two
        print(f"Multiply: {result}")

    @staticmethod
    def add(number_one, number_two):
        result = number_one + number_two
        print(f"Addition: {result}")
```

```python
Calculator.add(2, 5)
```

Notice here that we didn't need to first can access the method without creating an instance of the class. Let's try it with out ``multiply()`` method. In this case, it needs to be called by creating an instance because it uses the class properties.

```python
Object = Calculator(2,5)
Object.multiply()
```

### Example: Using the built-in staticmethod() function

We can also use the `staticmethod()` function to convert a function to a static function.

```python
class Person:
    """
    Class that defines a person who has an age defined by the age() method.
    """

    def age(age_number):
        if age_number <= 30:
            print("Young")

        elif age_number <= 50:
            print("Middle Age")

        else:
            print("Senior Age")


Person.age_category = staticmethod(Person.age)
John = Person()
John.age_category(45)
```
